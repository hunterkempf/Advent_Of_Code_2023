```python
def create_diff_list(vals_list):
    output_list = []
    for i in range(len(vals_list)-1):
        output_list.append(vals_list[i+1]-vals_list[i])
    return output_list

def all_zeros(vals_list):
    return sum(vals_list)==0
```

```python
grand_total = 0
for line in lines:
    vals_array = []
    values = line.replace("\n","").split(" ")
    values = [int(x) for x in values]
    vals_array.append(values)
    prev_line = values
    while(1):
        next_line = create_diff_list(prev_line)
        vals_array.append(next_line)
        if all_zeros(next_line):
            break
        prev_line=next_line
    print(vals_array)
    total =0
    for i in range(len(vals_array)-2,-1,-1):
        total += vals_array[i][-1]
    grand_total += total
grand_total
        

    
```
