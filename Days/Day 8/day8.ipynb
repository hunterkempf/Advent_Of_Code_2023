```python
instructions = lines[0].replace("\n","")
import re
data_map = {}
for line in lines[2:]:
    line_info = re.search(r"([0-9A-Z]*) = \(([0-9A-Z]*), ([0-9A-Z]*)\)",line)
    data_map[line_info.group(1)] = {"L":line_info.group(2),"R":line_info.group(3)}
data_map
```

```python
starting_keys = []
for key in data_map.keys():
    if key[2]=="A":
        starting_keys.append(key)
starting_keys
```

```python
# need to find cycle 
all_keys_Z_lists=[]
for starting_key in starting_keys:
    curr_node = starting_key
    count = 0
    end_Z_list = []
    for instruction in instructions*1000:
    #instruction = instructions[count%len(instructions)]
        curr_node = data_map[curr_node][instruction]
        count += 1
        if curr_node[2]=="Z":
            end_Z_list.append([curr_node,count])
    all_keys_Z_lists.append(end_Z_list)
all_keys_Z_lists
```

```python
from math import lcm
nums = []
for lista in all_keys_Z_lists:
    nums.append(int(lista[0][1]))
lcm(*nums)
```
