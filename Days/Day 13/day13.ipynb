```python
def find_row_reflection(pattern):
    guesses = []
    for i in range(1,len(pattern),1):
        if pattern[i]==pattern[i-1]:
            guesses.append(i)
    found_reflection = 0
    for guess in guesses:
        curr_top = guess-2
        curr_bot = guess+1
        while(1):
            if curr_bot>=len(pattern) or curr_top<0:
                found_reflection = guess
                break
            if pattern[curr_top] != pattern[curr_bot]:
                break
            else:
                curr_top -=1
                curr_bot +=1
    return found_reflection

def find_col_reflection(pattern):
    columns = []
    for i in range(len(pattern[0])):
        column = []
        for row in pattern:
            try:
                column.append(row[i])
            except:
                print(row)
                column.append(row[i])
        columns.append("".join(column))

    guesses = []
    for i in range(1,len(columns),1):
        if columns[i]==columns[i-1]:
            guesses.append(i)
    found_reflection = 0
    for guess in guesses:
        curr_top = guess-2
        curr_bot = guess+1
        while(1):
            if curr_bot>=len(columns) or curr_top<0:
                found_reflection = guess
                break
            if columns[curr_top] != columns[curr_bot]:
                break
            else:
                curr_top -=1
                curr_bot +=1
    return found_reflection
```

```python
patterns = []
curr_pattern = []

for line in lines:
    line = line.replace("\n","")
    if line == "":
        patterns.append(curr_pattern)
        curr_pattern=[]
    else:
        curr_pattern.append(line)
patterns.append(curr_pattern)
total=0
for pattern in patterns:
    row_reflection = find_row_reflection(pattern)
    total += row_reflection*100
    col_reflection = find_col_reflection(pattern)
    total += col_reflection
total
```
